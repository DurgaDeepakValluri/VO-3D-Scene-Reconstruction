# Visual Odometry + 3D Scene Reconstruction



## Aim:

To build a modular Visual Odometry pipeline that takes a monocular video as input and estimates the camera’s 6-DoF trajectory over time, while simultaneously reconstructing a sparse 3D point cloud of the scene using feature-based tracking and triangulation — all in Python using OpenCV and Open3D, with zero reliance on deep learning or ROS.
